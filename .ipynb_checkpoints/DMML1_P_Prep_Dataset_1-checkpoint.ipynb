### Refactoring categorical variables into fewer values

- Grouping 'workclass' variable into 6 categories.

```python
workclass_map = {' State-gov': ' Government', 
                 ' Self-emp-not-inc': ' Self-employed', 
                 ' Federal-gov': ' Government', 
                 ' Local-gov': ' Government', 
                 ' ?': ' Unknown', 
                 ' Self-emp-inc': ' Self-employed'}
```

```python
data['workclass'] = data['workclass'].replace(workclass_map)
```

```python
data.groupby(['workclass'])['workclass'].count().sort_values(ascending=False)
```

- Grouping 'education' variable into 5 categories.

```python
data['education'] = data['education'].replace(education_map)
```

```python
data.groupby(['education'])['education'].count().sort_values(ascending=False)
```

- Grouping 'marital-status' variable into 3 categories.

```python
marital_map = {' Never-married': ' Married',
               ' Married-civ-spouse': ' Married', 
               ' Married-spouse-absent': ' Married',
               ' Separated': ' Divorced', 
               ' Married-AF-spouse':' Married'}
```

### Balance of Dataset

- The dataset has 70% of the values 0 (<=50K) and 30% of values 1 (>50K). This **unbalanced** situation could impact the accuracy of the models so we are going to split the dataset more equally.

```python
data.groupby(['salary'])['salary'].count().sort_values(ascending=False)
```

- Plotting the balance of target variable (Salary)

```python
fig, ax = plt.subplots(figsize=(10,5))
ax.set_title('Balance of Target Variable (Salary)', fontsize=16)
sns.countplot(data['salary'],palette='hls')
plt.savefig(path + 'Figures/adult_balance_binary.png')
plt.show()
```

- Resampling the target variable (Salary)

```python
from sklearn.utils import resample
```

```python
data_0 = data[data.salary == 0]
data_1 = data[data.salary == 1]

# Downsample "majority" class

data_0_down = resample(data_0,
                      replace=False,
                      n_samples=7841,
                      random_state=123)

# Combine the down sampled dataset with the minority dataset

data_final = pd.concat([data_0_down,data_1])
```

- Result after **resampling** the target variable.

```python
data_final.groupby(['salary'])['salary'].count().sort_values(ascending=False)
```

- Plotting the balance of the dataset after resampling.

```python
fig, ax = plt.subplots(figsize=(10,5))
ax.set_title('Dataset Rebalanced (Salary)', fontsize=16)
sns.countplot(data_final['salary'],palette='hls')
plt.savefig(path + 'Figures/adult_balance_binary_new.png')
plt.show()
```
